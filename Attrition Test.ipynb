```python
began_treatment = [
    not math.isnan(x) for x in raw_data[
    ('NQ4.1_First Click', 'Timing - First Click')
    ]
]

began_control = [
    not math.isnan(x) for x in raw_data[
    ('NQ3.1_First Click', 'Timing - First Click')
    ]
]

began_survey = pd.concat([raw_data[began_treatment], raw_data[began_control]])
```

```python
began_survey['Completed Treatment'] = [
    (not math.isnan(x)) * 1 for x in began_survey[
    ('Q2.5_11', 'How much do you agree with this statement from 0-100? 0 being the lowest and 100 being the highest. Slide to enter a value. - Level of Agreement')
    ]
]

began_survey['Completed Control'] = [
    (not math.isnan(x)) * 1 for x in began_survey[
    ('Q1.5_0', 'How much do you agree with this statement from 0-100? 0 being the lowest and 100 being the highest. Slide to enter a value.\n\n\n\n\n"Undocumented workers are often exploited in the U.S., despite being a crucial part of the nation’s economy. To protect their rights and drive economic growth, we should take immediate steps to significantly expand the availability of work visas." - Level of Agreement')
    ]
]
```

```python
began_survey['Treatment'] == True
```

```python
began_survey['TreatAssign'] = [x * 1 for x in began_survey[('Treatment', 'Treatment')]]
began_survey['Attrition'] = 1- (began_survey['Completed Treatment'] + began_survey['Completed Control'])
began_survey = began_survey[['NQ2.4','NQ2.5', 'NQ2.6','NQ2.7','NQ2.8','NQ2.9','TreatAssign', 'Completed Treatment', 'Completed Control', 'Attrition']]
```

```python
treatment_map = {'NQ2.4':'party','NQ2.5':'leaning','NQ2.6':'gender','NQ2.7':'age',
                 'NQ2.8':'race','NQ2.9':'hispanic','NQ2.10':'citizenship',
                 'Q2.1_11':'Q1','Q2.2_11':'Q2','Q2.3_11':'Q3','Q2.4_1':'Q4',
                 'Q2.5_11':'Q5'}
began_survey.rename(columns=treatment_map, inplace=True)
```

```python
began_survey.columns = began_survey.columns.droplevel(1)
```

```python
covariates = ['party','leaning','gender','age','race','hispanic']
```

```python
# From I. Vaimberg
# turn covariates into one hot encoded columns
df_dummy = pd.get_dummies(began_survey[covariates])

one_hot_covariates = [#'party_Democrat',
 'party_Independent', 'party_Other',
'party_Republican',

'leaning_Conservative', 'leaning_Lean Conservative',
'leaning_Lean Liberal', 'leaning_Liberal', 'leaning_Moderate',
'leaning_Strongly Conservative',
 #'leaning_Strongly Liberal',

#'gender_man',
 'gender_woman',

#'age_18-24',
'age_25-34', 'age_35-44',
'age_45-54', 'age_55-64', 'age_65 or older',

'race_American Indian or Alaska Native', 'race_Asian',
'race_Black or African American',
'race_Native Hawaiian or Other Pacific Islander', 'race_Other race',
'race_Two or more',
#'race_White',

#'hispanic_No',
 'hispanic_Yes',
]


began_survey[one_hot_covariates] = df_dummy[one_hot_covariates]
```

```python
rs = began_survey


X = rs[['TreatAssign']+one_hot_covariates]#+interaction_cols]
y = rs['Attrition']


X = sm.add_constant(X)
basic_model = sm.OLS(y,X.astype(float))
basic_results = basic_model.fit(cov_type='HC3')
print(basic_results.summary())
```
